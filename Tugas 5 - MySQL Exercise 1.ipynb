## MySQL Exercise 1

### Sample data yang di gunakan 'classicmodels.sql'

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. Prepare a list of offices sorted by country, state, city.

```python
run("""
    SELECT country, state, city
    FROM offices 
    """)
```

### 2. How many employees are there in the company?

```python
run("SELECT COUNT(*) AS JumlahKaryawan FROM employees")
```

### 3. What is the total of payments received?

```python
run("SELECT SUM(amount) AS totalPayment FROM payments")
```

### 4. List the product lines that contain 'Cars'.

```python
run("SELECT productline FROM productlines WHERE productline LIKE '%cars%' ")
```

### 5. Report total payments for October 28, 2004.

```python
run("SELECT paymentDate, amount FROM payments WHERE paymentDate = '2004-10-28' ")
```

### 6. Report those payments greater than \$100,000

```python
run("SELECT amount FROM payments HAVING amount>100000")
```

### 7. List the products in each product line.

```python
run("""
SELECT pl.productLine, 
       pd.productName 
FROM productlines pl
INNER JOIN products pd
    ON pl.productline = pd.productline
""")
```

### 8. How many products in each product line?

```python
run("SELECT productLine, COUNT(productName) AS totalProducts FROM products GROUP BY productLine")
```

### 9. What is the minimum payment received?

```python
run("SELECT MIN(amount) FROM payments")
```

### 10. List all payments greater than twice the average payment.

```python
run(""" 
    SELECT * FROM Payments 
    WHERE amount > (SELECT AVG(amount)*2 FROM Payments)
    """)
```

### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT AVG((MSRP-buyPrice)/buyPrice*100) as averagePercentageMarkup FROM products")
```

### 12. How many distinct products does ClassicModels sell?

```python
run("SELECT COUNT(DISTINCT productName) AS totalProducts FROM products")
```

### 13. Report the name and city of customers who don't have sales representatives?

```python
run("SELECT customerName, city FROM customers WHERE salesRepEmployeeNumber IS NOT NULL")
```

### 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
run("""
SELECT 
    CONCAT(firstName, ' ', lastName) 
    AS fullName, jobTitle
FROM employees 
WHERE 
    jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

### 15. Which orders have a value greater than $5,000?

```python
run("""
SELECT od.orderNumber,
       od.orderDate,
       od.status,
       SUM(dt.quantityOrdered*dt.priceEach) AS totalPrice
FROM orders od
INNER JOIN orderdetails dt
    ON od.orderNumber = dt.orderNumber
GROUP BY dt.orderNumber
HAVING totalPrice>5000
""")
```
