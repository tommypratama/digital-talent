## Analisa Volume dan Nilai Ekspor Melalui DKI Jakarta Menurut Jenis Komoditi

Dataset ini berisi Volume dan Nilai Ekspor Melalui DKI Jakarta Menurut Jenis Komoditi pada tahun 2011-2012

Penjelasan mengenai Variabel pada Dataset ini:

1. tahun : Tahun
2. hs : Nomor HS
3. jenis_komoditas : Jenis komoditas ekspor
4. volume : Jumlah volume ekspor (dalam Ton)
5. nilai : Jumlah nilai ekspor (dalam ribu US$)

```python
import pandas as pd
import numpy as np
```

```python
ekspor_jakarta = pd.read_csv('volumedannilaiekspormelaluidkijakartamenurutjeniskomoditi20112012.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
ekspor_jakarta.sample(50)
```

#### Total Ekspor DKI Jakarta berdasarkan Jenis komoditas, Volume dan Nilai

```python
ekspor_jakarta.groupby(['jenis_komoditas','volume','nilai']).sum().sample(50)
```

#### Statistik Ekspor DKI Jakarta Tahun 2011 dan 2012

```python
print ('\n')
print ('\033[1m  Statistik Ekspor Tahun 2011  \033[0m','\n')
print('Jumlah Terendah Volume Ekspor Tahun 2011 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2011].volume.min())
print('Jumlah Tertinggi Volume Ekspor Tahun 2011 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2011].volume.max())
print('Komoditas Terendah Volume Ekspor Tahun 2011 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2011].jenis_komoditas.min())
print('Komoditas Tertinggi Volume Ekspor Tahun 2011 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2011].jenis_komoditas.max())
print('Jumlah Volume Ekspor Tahun 2011 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2011].volume.sum(), '\n')

print ('\033[1m  Statistik Ekspor Tahun 2012  \033[0m','\n')
print('Jumlah Terendah Volume Ekspor Tahun 2012 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2012].volume.min())
print('Jumlah Tertinggi Volume Ekspor Tahun 2012 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2012].volume.max())
print('Komoditas Terendah Volume Ekspor Tahun 2012 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2012].jenis_komoditas.min())
print('Komoditas Tertinggi Volume Ekspor Tahun 2012 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2012].jenis_komoditas.max())
print('Jumlah Volume Ekspor Tahun 2012 =', ekspor_jakarta[ekspor_jakarta['tahun'] == 2012].volume.sum(), '\n')

total=ekspor_jakarta.volume.sum()
print("Total Volume Ekspor DKI Jakarta ( Dalam Ton ) Tahun 2011-2012 =",total)
```

#### Visualisasi sederhana

```python
ax = ekspor_jakarta.groupby('tahun').mean()['volume']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Volume", fontsize=12)
```

```python
year_condition = ekspor_jakarta['tahun'] == 2011
ax = ekspor_jakarta[year_condition].groupby('jenis_komoditas').sum()['volume']
ax = ax.plot(kind='bar', title ="Jumlah Volume Berdasarkan Jenis Komoditas", figsize=(19,5), legend=True, fontsize=12)
ax.set_xlabel("Jenis Komoditas", fontsize=12)
ax.set_ylabel("Volume", fontsize=12)
```

```python
ax = ekspor_jakarta.groupby('nilai').sum()['volume']
ax = ax.plot(kind='line', title ="Jumlah Nilai Ekspor", figsize=(19,5), legend=True, fontsize=12)
ax.set_xlabel("Jenis Komoditas", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```

#### Analisa Data

Berdasarkan data visualisasi data disimpulkan bahwa, Volume Ekspor DKI Jakarta terjadi penurunan di Tahun 2012 dibanding Tahun 2011. Dan Komoditas Terendah Volume Ekspor Tahun 2011 yaitu Alas Kaki, Komoditas Tertinggi Volume Ekspor Tahun 2011 yaitu Wol, Bulu Hewan.
